```python
import matplotlib.pyplot as plt
import pandas as pd

# Data provided
data = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Publication_no": [28, 25, 46, 63, 61, 72, 56, 57, 84, 73],
    "Journal": [14, 14, 23, 24, 25, 43, 27, 34, 40, 42],
    "Conference": [14, 11, 23, 39, 36, 29, 29, 23, 44, 31]
}

df = pd.DataFrame(data)

# Create a stacked bar chart for publications split into Journals and Conference papers
plt.figure(figsize=(6, 3))

# Stacked bars for Journal and Conference
plt.bar(df['Year'], df['Journal'], label='Journal', color='lightblue', edgecolor='black')
plt.bar(df['Year'], df['Conference'], bottom=df['Journal'], label='Conference', color='orange', edgecolor='black')

# Adding labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Publications', fontsize=12)

plt.xticks(df['Year'], rotation=0)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the chart as a PDF
plt.tight_layout()
plt.savefig("Publications_Per_Year.pdf", format='pdf')
plt.show()

print("Chart has been saved as 'Publications_Per_Year.pdf'.")
```
